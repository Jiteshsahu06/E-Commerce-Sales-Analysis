# 1. univariate Analysis 

#### Ship Mode column

```python
data["Ship Mode"].value_counts()    # product are delivers to which ship mode
```

```python
x= data["Ship Mode"].value_counts().index  # this is x-axis 
print(x)
y=data["Ship Mode"].value_counts().values  # this is y-axis
print(y)
```

```python
plt.figure(figsize=(4,5))                                         
plt.pie(y,labels=x,startangle=50,autopct="%0.2f%%" ,radius=1.2 )  # find the shiping mode which is most used in dataset ?
plt.legend(x,loc=2)
plt.title("Count of Ship Mode")
plt.show()
```

####  Segment column

```python
data["Segment"].value_counts()  
```

```python
plt.figure(figsize=(4,3))                     # which type are customers Most Track in dataset ?
sns.countplot(x="Segment",data=data,hue="Segment",palette="hot")
plt.title("Count of Segment")
plt.show()
```

### How many sell on which Particular Year ? Find it.
